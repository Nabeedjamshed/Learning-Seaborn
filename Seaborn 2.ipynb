```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = np.linspace(1,5,12).reshape(3,4)
df
```

```python
sns.heatmap(df) # In heatmap we represent 2D data, use heatmap mostly for feature engineering
```

```python
globaldf = pd.read_csv(r"D:\Jupyter\Excel\global_warming.csv")
globaldf.head(3)
```

```python
# drop categorical data bcz heatmap numerical data leta hai, country ko drop nhi kya usse index bnadya take pta chale konse country ka data hai
globaldf = globaldf.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code']).set_index('Country Name')
globaldf
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globaldf, vmin=0, vmax=21) #vmin and vmax se values ko set krskte ke kha se kha tk ka data show krao
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm') #coolwarm is more good to understand the data 
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True) #annot se table mai jo data likha hai uski value bhi aajati ahi
```

```python
names = np.array([['a1','a2','a3','a4'],['a5','a6','a7','a8'],['a9','a10','a11','a12']])
sns.heatmap(df, cmap='viridis',annot=names, fmt='s') #by using annot wo values bhi dal skte hai or apni marzi se bhi values dalskte hai, fmt mai array ki type likhte hai s for string
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k') # linewidths=1,linecolor='k' this parameters set the line between the boxes and set ther colors
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', xticklabels=False, yticklabels=False, cbar=False)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', xticklabels=np.arange(0,15), #x and y axis ki values change bhi krskte hai
            yticklabels=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M','N', 'O'])
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'max'}
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'min'}
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'both'}
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'max',
            'orientation':'horizontal',
           'shrink':0.5,}
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'min',
            'extendfrac': 0.5,} #agr jitni nok barhani hai utna extendfrac ko barhado
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'min',
            'extendfrac': 0.1,
            'drawedges':True #isse edges draw hojate hai
           }
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {'extend':'min',
            'extendfrac': 0.1,
            'drawedges':True,
            'ticks':np.arange(0,22) #agr chate ho ke sb values show ho means 0 to 22 ki saare values show ho tu ticks use krlo
           }
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k', cbar_kws = cbar_kws)
```

```python
plt.figure(figsize=(16,9))
annot_kws = {'fontsize':12,
            'color':'k',
            'fontstyle':'italic'
            }
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k',annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
annot_kws = {'fontsize':10,
            'color':'k',
            'fontstyle':'italic',
             'backgroundcolor': 'w',
             'rotation': 'vertical',
             'verticalalignment':'center',
             'alpha':0.5
            }
sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k',annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
annot_kws = {'fontsize':10,
            'color':'k',
            'fontstyle':'italic',
            }
ax = sns.heatmap(globaldf, vmin=0, vmax=21, cmap='coolwarm', annot=True,linewidths=1,linecolor='k',annot_kws=annot_kws)
ax.set(title='Heatmap for Global warming',
       xlabel='Years',
       ylabel='Countries')
sns.set(font_scale=3) # ye sb font ka size parha dega
plt.show()
```

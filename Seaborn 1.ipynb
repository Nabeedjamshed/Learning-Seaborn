```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.lineplot(x='size', y='total_bill', data=df) #below graph show krrha ke jitne zyada people honge utna zyada bill dena parhega
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='sex') #hue nai jo bhi column ka name dete hai usmai jotni bhi values hoti hai ye uska graph plot krdeta hai
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day')
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day') #style hr line ko different style dedega matplotlib mai khud manually krna parhta hai
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='Accent_r') #palette se colormap ko change krskte hai
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='Blues_r')
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='CMRmap')
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='CMRmap',dashes=True, markers=['o','*','>','<']) #agr dashes lines nhi cheye tu dashes ko false krdo
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='CMRmap',dashes=True, markers=['o','*','>','<'],legend='brief')
```

```python
sns.set(style='whitegrid')
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='CMRmap',dashes=True, markers=['o','*','>','<'],legend='full')
```

```python
sns.set(style='darkgrid')
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='CMRmap',dashes=True, markers=['o','*','>','<'],legend='full')
```

```python
sns.set(style='dark')
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day',palette='CMRmap',dashes=True, markers=['o','*','>','<'],legend='full')
```

```python
sns.displot(df['tip'], kde=True)
```


In Seaborn, the distplot function is used to create a histogram combined with a Kernel Density Estimate (KDE) plot. The KDE plot is the smooth line that you see overlaid on the histogram.

```python
sns.displot(df['size'], kde=True)
```

```python
sns.displot(df['total_bill'], kde=True, color='r')
```

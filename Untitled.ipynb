# Catplot()
catplot ke zarie different plots ko plot krskte hai

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

```python
sns.catplot(x = 'age', y='alive', data=titanic_df, kind='box',orient='h',  hue='who') 
```

```python
sns.catplot(x = 'age', y='alive', data=titanic_df, kind='strip',orient='h',  hue='who')  #strip plot se easily pta lgrha hai ke man zyada survive nhi krpae while women zyada survive krgae
```

```python
sns.catplot(x = 'alive', data=titanic_df, kind='count',  hue='who') # count plot se alive mai yes and no ki occurence ko find kya tu isse easily pta chl gya ke 400+ men were died while around 200 women survive
```

```python
sns.catplot(x = 'age', y='alive', data=titanic_df, kind='boxen', hue='who')
```

```python
sns.catplot(x = 'age', y='alive', data=titanic_df, kind='violin', hue='who')
```
